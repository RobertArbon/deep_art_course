# Audio (Eurasian Wren)


```python
import librosa
import librosa.display
```

```python
waveform, sampling_rate = librosa.load('audio/XC248037 - Eurasian Wren - Troglodytes troglodytes.mp3')
```

```python
sampling_rate
```

```python
waveform
```

```python
display.Audio(data=waveform, rate=sampling_rate)
```

## Amplitude vs time

```python
time = np.arange(waveform.shape[0])/float(sampling_rate)
plt.scatter(time, waveform, s=2, alpha=0.5)
plt.xlabel('time (seconds)')
plt.ylabel('amplitude')
plt.savefig('images/wren_waveform.png')
```

## Frequencies and amplitdue vs time (spectrogram/sonogram)
### All of the audio

```python
S = librosa.feature.melspectrogram(waveform, sr=sampling_rate, n_mels=128)
log_S = librosa.power_to_db(S, ref=np.max)
plt.figure(figsize=(12,4))
librosa.display.specshow(log_S, sr=sampling_rate, x_axis='time', y_axis='mel')
plt.title('mel power spectrogram')
plt.colorbar(format='%+02.0f dB')
plt.tight_layout()
```

### First 10 seconds

```python
S = librosa.feature.melspectrogram(waveform[:10*sampling_rate], sr=sampling_rate, n_mels=128)
log_S = librosa.power_to_db(S, ref=np.max)
plt.figure(figsize=(12,4))
librosa.display.specshow(log_S, sr=sampling_rate, x_axis='time', y_axis='mel')
plt.title('mel power spectrogram')
plt.colorbar(format='%+02.0f dB')
plt.tight_layout()
plt.savefig('images/wren_sonogram.png')
```
